# **Traffic Object Detection + Object Distance Estimation**

```python
def plotDot(x, y, img, color='grey'):
    if color == 'green':
        rgb = (125,218,140)
    else:
        rgb = (178, 178, 178)
    pinnedImg = cv2.circle(img, (x,y), radius=15, color=rgb, thickness=-1)
    return pinnedImg


def plotTriangles(img):
    height, width, channels = img.shape

    lt = []
    ct = []
    rt = []

    # Center Triangle Coordinates (3 points)
    ct.append( round((width / 2) * 0.55) )                   # Center Triangle Left X
    ct.append( round(height * 0.97)  )                       # Center Triangle Left Y

    ct.append( round(width / 2) )                            # Center Triangle Top X
    ct.append( round((height / 2) * 1.1) )                   # Center Triangle Top Y

    ct.append( round((width / 2) + ((width / 2) * 0.55)) )   # Center Triangle Right X
    ct.append( round(height * 0.97)  )                       # Center Triangle Right Y


    # Left Triangle Coordinates (3 points)
    lt.append(0)                                 # Left Triangle Left X
    lt.append( round((height / 2) * 1.5)  )      # Left Triangle Left Y

    lt.append( round(width / 2) )                # Left Triangle Top X
    lt.append( round((height / 2) * 1.1) )       # Left Triangle Top Y

    lt.append( round((width / 2) * 0.55) )       # Left Triangle Right X
    lt.append( round(height * 0.97)  )           # Left Triangle Right Y


    # Right Triangle Coordinates (3 points)
    rt.append( round((width / 2) + ((width / 2) * 0.55)) )   # Right Triangle Left X
    rt.append( round(height * 0.97)  )                       # Right Triangle Left Y

    rt.append( round(width / 2) )                            # Right Triangle Top X
    rt.append( round((height / 2) * 1.1) )                   # Right Triangle Top Y

    rt.append(width)                                         # Right Triangle Right X
    rt.append( round((height / 2) * 1.5)  )                  # Right Triangle Right Y

    return lt, ct, rt



def point_in_triangle(point, triangle):

    # Unpack arguments
    x, y = point
    ax = triangle[0]
    ay = triangle[1]
    bx = triangle[2]
    by = triangle[3]
    cx = triangle[4]
    cy = triangle[5]
    # Segment A to B
    side_1 = (x - bx) * (ay - by) - (ax - bx) * (y - by)
    # Segment B to C
    side_2 = (x - cx) * (by - cy) - (bx - cx) * (y - cy)
    # Segment C to A
    side_3 = (x - ax) * (cy - ay) - (cx - ax) * (y - ay)
    # All the signs must be positive or all negative
    return (side_1 < 0.0) == (side_2 < 0.0) == (side_3 < 0.0)



def calc_Distance(point, img):
    height, width, channels = img.shape
    x, y = point
    
    roadLengthMin = round((height / 2) * 1.1)
    roadLengthMax = round(height * 0.75)
    
    closeness = ( y - roadLengthMin ) / ( roadLengthMax -roadLengthMin )
    
    return closeness



def calc_BC_Points(segmask):
    bc_points = []
    for i in range(len(segmask['rois'])):
        bc_point= []
        box = segmask["rois"][i]
        
        x = round( (box[1] + box[3]) / 2 )
        y = box[2]
        bc_point.extend([x, y])
        bc_points.append(bc_point)
        
    return bc_points



def plot_PointsInLane(bc_points, img):
    
    radarImg = cv2.imread("/content/drive/MyDrive/Colab Notebooks/VideoObjectSegmentationProject/CarRadarGraphicsExtended.png")
    radarMax = 450
    
    lt, ct, rt = plotTriangles(img)

    for p in bc_points:
        x = p[0]
        y = p[1]
        #print("Processing "+str(x)+", "+str(y)+": ")
        
        point = (x, y)
        check = point_in_triangle(point, lt)
        if check == True:
            #print("lt check = "+str(check))
            closeness = calc_Distance(point, img)
            x = 47
            y = round(radarMax * closeness)
            radarImg = plotDot(x, y, radarImg)
            
        check = point_in_triangle(point, ct)
        if check == True:
            #print("ct check = "+str(check))
            closeness = calc_Distance(point, img)
            x = 117
            y = round(radarMax * closeness)
            radarImg = plotDot(x, y, radarImg, color = 'green')
            
        check = point_in_triangle(point, rt)
        if check == True:
            #print("rt check = "+str(check))
            closeness = calc_Distance(point, img)
            x = 187
            y = round(radarMax * closeness)
            radarImg = plotDot(x, y, radarImg)
            
    return radarImg



def segmentFrame_Driver(imgpath):
    
    instance_seg = instance_segmentation()
    instance_seg.load_model("/content/drive/MyDrive/Colab Notebooks/VideoObjectSegmentationProject/mask_rcnn_coco.h5")
    target_classes = instance_seg.select_target_classes(person=True, car=True)
    segmask, output = instance_seg.segmentImage(imgpath, segment_target_classes= target_classes, show_bboxes= True)
    #cv2.imwrite("segOutput.jpg", output)

    img = output
    bc_points = calc_BC_Points(segmask)
    radarImg = plot_PointsInLane(bc_points, img)
    #cv2.imwrite("radarOutput.jpg", radarImg)

    # merge video frame with radar frame
    outputmerge = output.copy()
    x_offset=round(outputmerge.shape[1] * 0.85)
    y_offset=50
    scale_percent = 70 # percent of original size
    width = int(radarImg.shape[1] * scale_percent / 100)
    height = int(radarImg.shape[0] * scale_percent / 100)
    dim = (width, height)
    radarImgResized = cv2.resize(radarImg, dim, interpolation = cv2.INTER_AREA)

    outputmerge[ y_offset:y_offset+radarImgResized.shape[0], x_offset:x_offset+radarImgResized.shape[1] ] = radarImgResized
    #cv2.imwrite("mergeOutput.jpg", outputmerge)
    
    return outputmerge



def checkFrameSize(img):
    height, width, channels = img.shape
    size = (width,height)
    return size

```

```python
# Creating a VideoCapture object to read the video
#cap = cv2.VideoCapture('dashcamVideoSample.mp4')
lst = []

vidcap = cv2.VideoCapture('/content/drive/MyDrive/Colab Notebooks/VideoObjectSegmentationProject/dashcamVideoSample.mp4')

success,image = vidcap.read()
count = 0
success = True
frame_array = []
while success:
    success,image = vidcap.read()
    if success == False:
        break
    if count == 0:
        size = checkFrameSize(image)
    
    cv2.imwrite("p1.jpg", image)
    outputframe = segmentFrame_Driver("p1.jpg")
    
    frame_array.append(outputframe)
    
    print ('Frame '+str(count+1))
    count += 1
    
out = cv2.VideoWriter('finalOutput.avi',cv2.VideoWriter_fourcc(*'DIVX'), 60, size)
for i in range(len(frame_array)):
    out.write(frame_array[i])


```
